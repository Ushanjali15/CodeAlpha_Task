```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from io import StringIO

# === STEP 1: Create sample unemployment data ===
data = StringIO("""
Date,Unemployment_Rate
2018-01-01,4.5
2018-02-01,4.6
2018-03-01,4.4
2019-01-01,5.0
2019-02-01,5.1
2019-03-01,5.2
2020-01-01,7.2
2020-02-01,7.5
2020-03-01,15.0
2020-04-01,23.5
2020-05-01,20.9
2021-01-01,10.0
2021-06-01,7.3
2022-01-01,6.2
2023-01-01,5.5
""")

# === STEP 2: Load data ===
df = pd.read_csv(data)
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

# === STEP 3: Data overview ===
print("Summary Statistics:\n", df.describe())
print("\nMissing Values:\n", df.isnull().sum())

# === STEP 4: Unemployment Trend Over Time ===
plt.figure(figsize=(12,6))
sns.lineplot(x='Date', y='Unemployment_Rate', data=df, label='Unemployment Rate')
plt.axvspan(pd.to_datetime('2020-03-01'), pd.to_datetime('2021-03-01'),
            color='red', alpha=0.2, label='COVID-19 Period')
plt.title("Unemployment Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# === STEP 5: Yearly Average Unemployment ===
yearly_avg = df.groupby('Year')['Unemployment_Rate'].mean().reset_index()

plt.figure(figsize=(10,5))
sns.barplot(x='Year', y='Unemployment_Rate', data=yearly_avg, palette='magma')
plt.title("Yearly Average Unemployment Rate")
plt.ylabel("Average Rate (%)")
plt.xlabel("Year")
plt.tight_layout()
plt.show()

# === STEP 6: Monthly Pattern (Seasonality) ===
monthly_avg = df.groupby('Month')['Unemployment_Rate'].mean().reset_index()

plt.figure(figsize=(10,5))
sns.lineplot(x='Month', y='Unemployment_Rate', data=monthly_avg, marker='o')
plt.title("Seasonal Trend: Monthly Average Unemployment")
plt.xlabel("Month")
plt.ylabel("Average Rate (%)")
plt.xticks(range(1, 13))
plt.grid(True)
plt.tight_layout()
plt.show()

# === STEP 7: COVID-19 Impact ===
pre_covid_avg = df[df['Date'] < '2020-03-01']['Unemployment_Rate'].mean()
post_covid_avg = df[df['Date'] >= '2020-03-01']['Unemployment_Rate'].mean()

print(f"\nPre-COVID Avg Unemployment: {pre_covid_avg:.2f}%")
print(f"Post-COVID Avg Unemployment: {post_covid_avg:.2f}%")

if post_covid_avg > pre_covid_avg:
    print(" COVID-19 had a significant negative impact on unemployment.")
else:
    print(" Unemployment didn't increase significantly post-COVID.")

# === STEP 8: Policy Insight Suggestion ===
print("\n Policy Insights:")
print("- Emergency relief programs are essential during pandemics or crises.")
print("- Upskilling and job transition programs should be prioritized during recovery.")
print("- Monitor seasonal unemployment (e.g. early-year spikes) and plan temporary support.")
```
